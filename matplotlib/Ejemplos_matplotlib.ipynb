```python
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots(figsize=(8, 6))

datos = [34.3, 24.3, 17.9, 9.7, 13.6]
etiquetas = ['PP', 'PSOE', 'VOX', 'CHA', 'Otros']
colores = ['#3498db', '#e74c3c', '#2ecc71', '#7e2c39', '#808080']


# El truco: wedgeprops con width < 1
ax.pie(datos,
       labels=None,
       colors=colores,
       autopct='%1.1f%%',
       startangle=90,
       wedgeprops={'width': 0.6,       # Grosor del anillo
                   'edgecolor': 'white',
                   'linewidth': 2},
       pctdistance=0.8,
       textprops={'fontsize': 14})               # % más afuera

ax.legend(etiquetas,
          title="Partidos políticos",
          loc="center left",
          bbox_to_anchor=(1, 0, 0.5, 1))

# Texto central
ax.text(0, 0, 'Total\n100%',
        ha='center', va='center',
        fontsize=12, fontweight='bold')

ax.set_title('Resultados Elecciones Aragon 2026')

plt.show()
```

```python
# Datos
meses = ['Ene', 'Feb', 'Mar', 'Abr', 'May', 'Jun',
         'Jul', 'Ago', 'Sep', 'Oct', 'Nov', 'Dic']
temp_madrid = [6.3, 7.9, 11.2, 13.7, 17.4, 22.8,
               26.2, 25.6, 21.1, 15.2, 9.8, 6.9]
temp_jaen = [10.4, 12.1, 16.2, 19.3, 24.8, 30.6,
              34.1, 33.5, 28.7, 22.1, 15.4, 11.2]

# TODO: Crear figura y axes con plt.subplots()
fig, ax = plt.subplots(figsize=(10, 6))

# TODO: Dibujar línea con ax.plot() - incluir marcadores
ax.plot(meses, temp_madrid, marker='o', label="Madrid", linewidth=2,
        color='#3498db', markersize=8)
ax.plot(meses, temp_jaen, marker='s', label="Jaen" ,linewidth=2,
        color='#e74c3c', markersize=8)

# TODO: Añadir título descriptivo
ax.set_title('Temperaturas Medias Mensuales', fontsize=16, fontweight='bold')


# TODO: Añadir etiquetas de ejes (con unidad en Y)
ax.set_xlabel('Meses', fontsize=12)
ax.set_ylabel('Temperaturas (ºC)', fontsize=12)

# TODO: Añadir grid
ax.grid(True, alpha=0.3)

ax.legend(title="Partidos políticos",
          loc="center left",
          bbox_to_anchor=(1, 0, 0.5, 1))

# TODO: Usar tight_layout()
plt.tight_layout()
plt.show()
```
